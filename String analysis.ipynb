## String Analysis

```python
#len()sed for calculating the length of a seq
protein_seq1 = 'VLSPADKTNVK'
len(protein_seq1)
```

```python
#Splicing python indexing starts from 0,1,2,...n
protein_seq1[3]
```

```python
#Splicing a subpart needs to be +1 at the end of the end number
protein_seq1[2:(5+1)]
```

```python
#Skipping a splicing
protein_seq1[1:(5+1):1]
#Changing the step size
protein_seq1[1:(5+1):2]
protein_seq1[1:(11+1):2]
protein_seq1[2:(5+1):2]
protein_seq1[2:(11+1):2]
protein_seq1[2:(11+1):1]
```

```python
protein_seq1[-1]
protein_seq1[::-1] #Reversing a sequence or flipping
```

```python
#Conacatenation (Between same data type only)
dna1 = 'AAT'
dna2 = 'TTA'
full = dna1 + dna2
full
```

```python
#Concatenation and formatting
rep_seq = 'GGGGCC'
rep_len = 4
actual_seq ='(' + rep_seq + ')' + str(rep_len) #Approach one
actual_seq

actual_seq2 = f"({rep_seq}){str(rep_len)}"
actual_seq2
```

```python
#Changing the case of a given string
#.lower() for upper case to lower case
#.upper() vice-versa
variable_name = 'abcd efgh'
variable_name.upper()

variable_name.title() # caps every first letter of evey word
variable_name.capitalize() # first letter caps of the string
```

```python
# .replace() used to create mutants in wild type
#variable_name.replace('a','b') changes a wiht b in variable_name
protein1 = 'VLSPADKTNVWWWYY'
mutant1 = protein1.replace('W','M') #changes W with M
print(mutant1)
mutant2 = protein1.replace('W','M',1) #replaces only first occurrence with M changing the step size to 2,3,4.. replaces accordingly
print(mutant2)
```

```python
# .find() and .count()
#.find() only finds the first occurrence of what is being searched 
protein1.find('W') #the output shows the index number if its in a string, where the first occurence has shown
abc = 'i love music music'
abc.find('music')# inputs in python indexing
abc.count('music')
protein1.find('Z')
```

## SPLIT AND JOIN


```python
#.split() defaultly considers space while splitting the string into its elements and withtin the splitted list the elements are separated by commas
str1 = 'ab, dcd efss hssssi'
atoz = str1.split()
print(atoz)
atoz2 = str1.split(',')
print(atoz2)

#.join()
atozrecombine = '/'.join(atoz)
print(atozrecombine)
```
